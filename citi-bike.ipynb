```python
# Import necessary libraries
import pandas as pd
import os
```

```python
# Step 1: Load the CSV files by specifying their paths manually
file1 = "september_2024/202409-citibike-tripdata_1.csv"
file2 = "september_2024/202409-citibike-tripdata_2.csv"
file3 = "september_2024/202409-citibike-tripdata_3.csv"
file4 = "september_2024/202409-citibike-tripdata_4.csv"
file5 = "september_2024/202409-citibike-tripdata_5.csv"
```

```python
# Step 2: Read each CSV file into a separate DataFrame
df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)
df3 = pd.read_csv(file3)
df4 = pd.read_csv(file4)
df5 = pd.read_csv(file5)
```

```python
# Step 3: Merge all DataFrames into one
merged_df = pd.concat([df1, df2, df3, df4, df5], ignore_index=True)
```

```python
merged_df.head()
```

```python
# Step 4: Clean the dataset
# Drop duplicate rows
merged_df.drop_duplicates(inplace=True)
# Handling missing values
cleaned_df = merged_df.dropna()
```

```python
cleaned_df.head()
```

```python
# Step 5: Export the cleaned dataset as a CSV file
output_path = "/Users/ellengrant/Desktop/Citi-Bike/output"
# Check if the directory exists, and create it if it doesn't
if not os.path.exists(output_path):
    os.makedirs(output_path)
# Full file path
output_file_path = os.path.join(output_path, "cleaned_data.csv")

cleaned_df.to_csv(output_file_path, index=False)
```
